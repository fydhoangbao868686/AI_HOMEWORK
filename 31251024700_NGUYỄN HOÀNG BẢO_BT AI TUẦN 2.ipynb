## Bài 2.11

```python
Ride_distance = ctrl.Antecedent(np.arange(0, 51, 1), 'Ride_distance')
Traffic_condition = ctrl.Antecedent(np.arange(0, 101, 1), 'Traffic_condition')
Demand_level = ctrl.Antecedent(np.arange(0, 101, 1), 'Demand_level')
Weather_condition = ctrl.Antecedent(np.arange(0, 3.1, 0.1), 'Weather_condition')
Customer_rating = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'Customer_rating')
Ride_punctuality = ctrl.Antecedent(np.arange(0, 101, 1), 'Ride_punctuality')
Customer_fare = ctrl.Consequent(np.arange(0, 1010000, 1000), 'Customer_fare')
Driver_point = ctrl.Consequent(np.arange(0, 100, 1), 'Driver_point')
Ride_distance['S'] = fuzz.trimf(Ride_distance.universe, [0, 1.5, 3])
Ride_distance['M'] = fuzz.trimf(Ride_distance.universe, [2, 5, 8])
Ride_distance['L'] = fuzz.trimf(Ride_distance.universe, [6, 13, 20])
Ride_distance['VL'] = fuzz.trimf(Ride_distance.universe, [15, 32.5, 50])
Traffic_condition['L'] = fuzz.trimf(Traffic_condition.universe, [0, 15, 30])
Traffic_condition['M'] = fuzz.trimf(Traffic_condition.universe, [20, 50, 70])
Traffic_condition['H'] = fuzz.trimf(Traffic_condition.universe, [60, 90, 100])
Demand_level['L'] = fuzz.trimf(Demand_level.universe, [0, 15, 30])
Demand_level['M'] = fuzz.trimf(Demand_level.universe, [20, 50, 70])
Demand_level['H'] = fuzz.trimf(Demand_level.universe, [60, 90, 100])
Weather_condition['Good'] = fuzz.trimf(Weather_condition.universe, [0, 1, 1.5])
Weather_condition['Moderate'] = fuzz.trimf(Weather_condition.universe, [1, 2, 2.5])
Weather_condition['Bad'] = fuzz.trimf(Weather_condition.universe, [2.25, 2.75, 3 ])
Customer_rating['L'] = fuzz.trimf(Customer_rating.universe, [0, 1, 2])
Customer_rating['M'] = fuzz.trimf(Customer_rating.universe, [1.5, 2.75, 4])
Customer_rating['H'] = fuzz.trimf(Customer_rating.universe, [4, 5, 5])
Ride_punctuality['Late'] = fuzz.trimf(Ride_punctuality.universe, [0, 25, 50])
Ride_punctuality['On_time'] = fuzz.trimf(Ride_punctuality.universe, [40, 60, 80])
Ride_punctuality['Early'] = fuzz.trimf(Ride_punctuality.universe, [70, 100, 100])
Customer_fare['Low'] = fuzz.trimf(Customer_fare.universe, [0, 15000, 25000])
Customer_fare['Medium'] = fuzz.trimf(Customer_fare.universe, [25000, 37500,50000])
Customer_fare['High'] = fuzz.trimf(Customer_fare.universe, [50000, 65000, 80000])
Customer_fare['Very High'] = fuzz.trimf(Customer_fare.universe, [75000, 100000, 100000])
Driver_point['None'] = fuzz.trimf(Driver_point.universe, [0, 7, 15])
Driver_point['Few'] = fuzz.trimf(Driver_point.universe, [15, 35,50])
Driver_point['Moderate'] = fuzz.trimf(Driver_point.universe, [50, 65, 80])
Driver_point['High'] = fuzz.trimf(Driver_point.universe, [80, 95, 100])
rule1 = ctrl.Rule(Customer_rating['L'] & Ride_punctuality['Late'],Driver_point['None'] )
rule2 = ctrl.Rule(Customer_rating['L'] & Ride_punctuality['On_time'],Driver_point['Few'] )
rule3 = ctrl.Rule(Customer_rating['L'] & Ride_punctuality['Early'],Driver_point['Few'] )
rule4 = ctrl.Rule(Customer_rating['M'] & Ride_punctuality['Late'],Driver_point['Few'] )
rule5 = ctrl.Rule(Customer_rating['M'] & Ride_punctuality['On_time'],Driver_point['Moderate'] )
rule6 = ctrl.Rule(Customer_rating['M'] & Ride_punctuality['Early'],Driver_point['Moderate'] )
rule7 = ctrl.Rule(Customer_rating['H'] & Ride_punctuality['Late'],Driver_point['Moderate'] )
rule8 = ctrl.Rule(Customer_rating['H'] & Ride_punctuality['On_time'],Driver_point['High'] )
rule9 = ctrl.Rule(Customer_rating['H'] & Ride_punctuality['Early'],Driver_point['High'] )
rule10 = ctrl.Rule(Ride_distance['S'] & Traffic_condition['L'] & Demand_level['L'], Customer_fare['Low'])
rule11 = ctrl.Rule(Ride_distance['S'] & Traffic_condition['M'] & Demand_level['H'], Customer_fare['Medium'])
rule12 = ctrl.Rule(Ride_distance['M'] & Traffic_condition['H'] & Demand_level['H'], Customer_fare['Very High'])
rule13 = ctrl.Rule(Ride_distance['L'] & Traffic_condition['M'] & Weather_condition['Good'], Customer_fare['High'])
rule14 = ctrl.Rule(Ride_distance['L'] & Traffic_condition['H'] & Weather_condition['Moderate'], Customer_fare['High'])
rule15 = ctrl.Rule(Ride_distance['VL'] & Traffic_condition['H'] & Demand_level['H'], Customer_fare['Very High'])
rule16 = ctrl.Rule(Ride_distance['M'] & Traffic_condition['L'] & Demand_level['L'], Customer_fare['Medium'])
rule17 = ctrl.Rule(Ride_distance['S'] & Traffic_condition['H'] & Weather_condition['Bad'], Customer_fare['High'])
rule18 = ctrl.Rule(Ride_distance['VL'] & Weather_condition['Bad'], Customer_fare['Very High'])
rule19 = ctrl.Rule(Ride_distance['M'] & Traffic_condition['M'] & Weather_condition['Moderate'], Customer_fare['Medium'])
rule20 = ctrl.Rule(Ride_distance['M'] & Traffic_condition['H'] & Demand_level['M'], Customer_fare['High'])
control_system = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20])
ride_system = ctrl.ControlSystemSimulation(control_system)
ride_system.input['Ride_distance'] = 32.5
ride_system.input['Traffic_condition'] = 90
ride_system.input['Demand_level'] = 90
ride_system.input['Weather_condition'] = 2.75
ride_system.input['Customer_rating'] = 5
ride_system.input['Ride_punctuality'] = 85
ride_system.compute()
print ('Giá đi xe', f"{ride_system.output['Customer_fare']:.2f}")
print ('Điểm khách hàng', f"{ride_system.output['Driver_point']:.2f}" )
```

## Bài 2.12

```python
Store_rating = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'Store_rating')
Sales_volume = ctrl.Antecedent(np.arange(0, 10001, 1), 'Sales_volume')
Profit_margin = ctrl.Antecedent(np.arange(0, 101, 1), 'Profit_margin')
Seasonal_event = ctrl.Antecedent(np.arange(0, 11, 1), 'Seasonal_event')
Competitor_discounts = ctrl.Antecedent(np.arange(0, 71, 1), 'Competitor_discounts')
Discount_pct = ctrl.Consequent(np.arange(0, 71, 1), 'Discount_pct')
Store_rating ["Low"]= fuzz.trimf(Store_rating.universe, [0, 2, 4])
Store_rating ["Medium"]= fuzz.trimf(Store_rating.universe, [4, 4.25, 4.5])
Store_rating ["High"]= fuzz.trimf(Store_rating.universe, [4.5, 5, 5])
Sales_volume ["Low"]= fuzz.trimf(Sales_volume.universe, [0, 500, 1500])
Sales_volume ["Medium"]= fuzz.trimf(Sales_volume.universe, [1000, 4000, 7000])
Sales_volume ["High"]= fuzz.trimf(Sales_volume.universe, [7000, 10000, 10000])
Profit_margin ["Low"]= fuzz.trimf(Profit_margin.universe, [0, 15, 35])
Profit_margin ["Medium"]= fuzz.trimf(Profit_margin.universe, [25, 50, 75])
Profit_margin ["High"]= fuzz.trimf(Profit_margin.universe, [65, 100, 100])
Seasonal_event ["None"]= fuzz.trimf(Seasonal_event.universe, [0, 2, 2])
Seasonal_event ["Moderate"]= fuzz.trimf(Seasonal_event.universe, [1, 6, 6])
Seasonal_event ["High"]= fuzz.trimf(Seasonal_event.universe, [6, 10, 10])
Competitor_discounts ["Low"]= fuzz.trimf(Competitor_discounts.universe, [0, 15, 30])
Competitor_discounts ["Medium"]= fuzz.trimf(Competitor_discounts.universe, [25, 35, 50])
Competitor_discounts ["High"]= fuzz.trimf(Competitor_discounts.universe, [45, 60, 70])
Discount_pct ["Very Low"]= fuzz.trimf(Discount_pct.universe, [0, 5, 5])
Discount_pct ["Low"]= fuzz.trimf(Discount_pct.universe, [5, 10, 10])
Discount_pct ["Medium"]= fuzz.trimf(Discount_pct.universe, [10, 20, 20])
Discount_pct ["High"]= fuzz.trimf(Discount_pct.universe, [20, 40, 40])
Discount_pct ["Very High"]= fuzz.trimf(Discount_pct.universe, [40, 70, 70])
rule1 = ctrl.Rule(Store_rating['High'] & Sales_volume['High'] & Profit_margin['High'] , Discount_pct['Very Low'])
rule2 = ctrl.Rule(Store_rating['Low'] & Sales_volume['Low'] & Profit_margin['High'] , Discount_pct['High'])
rule3 = ctrl.Rule(Seasonal_event['High'] & Competitor_discounts ["High"] ,Discount_pct ["Very High"])
rule4 = ctrl.Rule(Store_rating ["Medium"] & Sales_volume['Medium'] & Profit_margin ["Medium"], Discount_pct ["Medium"])
rule5 = ctrl.Rule(Competitor_discounts ["Low"] & Profit_margin ["Low"] & Sales_volume ["High"], Discount_pct ["Very Low"])
rule6 = ctrl.Rule(Store_rating ["Low"] & Seasonal_event ["None"], Discount_pct ["Medium"])
rule7 = ctrl.Rule(Sales_volume ["Low"] & Profit_margin ["Low"], Discount_pct ["Very High"])
rule_control = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
discount_system = ctrl.ControlSystemSimulation(rule_control)
discount_system.input['Store_rating'] = 4.25
discount_system.input['Sales_volume'] = 5000
discount_system.input['Profit_margin'] = 20
discount_system.input['Seasonal_event'] = 7
discount_system.input['Competitor_discounts'] = 69
discount_system.compute()
print("Phần trăm giảm giá là", f"{discount_system.output['Discount_pct']:.2f}")
```

## Bài 2.13

```python
PD = ctrl.Antecedent(np.arange(0,101,1), 'Product_Demand')
CPP = ctrl.Antecedent(np.arange(0,11,1), 'Competitor_pricing_pressure')
SR= ctrl.Antecedent(np.arange(0,5.1,0.1), 'Store_Reputation')
PM = ctrl.Antecedent(np.arange(0,101,1), 'Profit_margin')
SD = ctrl.Antecedent(np.arange(0,11,1), 'Seasonal_demand')
Dis_pct = ctrl.Consequent(np.arange(0,71,1), 'Discount_pct')
PD['Low'] = fuzz.trimf(PD.universe, [0, 25, 30])
PD['Medium'] = fuzz.trimf(PD.universe, [20,50, 70])
PD['High'] = fuzz.trapmf(PD.universe, [60,80, 99, 100])
CPP['Low'] = fuzz.trimf(CPP.universe, [0, 1, 1])
CPP['Medium'] = fuzz.trapmf(CPP.universe, [1,2,5,7])
CPP['High'] = fuzz.trimf(CPP.universe, [6, 9, 10])
SR['Low'] = fuzz.trimf(SR.universe, [0, 1,4])
SR['Medium'] = fuzz.trapmf(SR.universe, [4,4,4.5,4.5])
SR['High'] = fuzz.trimf(SR.universe, [4.5, 4.9, 5])
PM['Low'] = fuzz.trimf(PM.universe, [0, 10, 25])
PM['Medium'] = fuzz.trimf(PM.universe, [20, 30, 40])
PM['High'] = fuzz.trimf(PM.universe, [ 30, 60, 70])
SD['Low'] = fuzz.trimf(SD.universe, [0, 1, 2])
SD['Medium'] = fuzz.trapmf(SD.universe, [1.5, 4, 5,7])
SD['High'] = fuzz.trimf(SD.universe, [6, 9, 10])
Dis_pct ['Very Low'] = fuzz.trimf(Dis_pct.universe, [0, 5, 5])
Dis_pct ['Low'] = fuzz.trapmf(Dis_pct.universe, [4,5,9,10])
Dis_pct ['Medium'] = fuzz.trapmf(Dis_pct.universe, [9, 10, 19, 20])
Dis_pct ['High'] = fuzz.trapmf(Dis_pct.universe, [19, 20, 40, 60])
Dis_pct ['Very High'] = fuzz.trimf(Dis_pct.universe, [50, 65, 70])
rule1 = ctrl.Rule(PD['High'] & CPP['Low'] & PM['Low'], Dis_pct['Very Low'])
rule2 = ctrl.Rule(PD['Low'] & CPP['High'] & PM['High'], Dis_pct['High'])
rule3 = ctrl.Rule(SR['High'] & PM['Medium'] & SD['High'], Dis_pct['Medium'])
rule4 = ctrl.Rule(CPP['High'] & SD['High'] & PM['High'], Dis_pct['Very High'])
rule5 = ctrl.Rule(SR['Low'] & PD['Medium'] & PM['Low'], Dis_pct['Medium'])
rule6 = ctrl.Rule(PD['High'] & SD['Low'] & CPP['Low'], Dis_pct['Very Low'])
rule7 = ctrl.Rule(PM['High'] & CPP['Medium'], Dis_pct['Medium'])
rule_control = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
discount_percentage = ctrl.ControlSystemSimulation(rule_control)
discount_percentage.input['Product_Demand'] = 80
discount_percentage.input['Competitor_pricing_pressure'] = 4
discount_percentage.input['Store_Reputation'] = 4.2
discount_percentage.input['Profit_margin'] = 60
discount_percentage.input['Seasonal_demand'] = 9
discount_percentage.compute()
print("Phần trăm giảm giá là", f"{discount_percentage.output['Discount_pct']:.2f}")
```

## Bài 2.14

```python
order_density = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'order_density')
deliver_urgency = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'deliver_urgency')
driver_load = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'driver_load')
traffic_condition = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'traffic_condition')
profit_per_delivery = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'profit_per_delivery')
combine_orders = ctrl.Consequent(np.arange(0, 11, 1), 'combine_orders')
delivery_priority = ctrl.Consequent(np.arange(0, 11, 1), 'delivery_priority')
order_density['low'] = fuzz.trimf(order_density.universe, [0, 0, 4])
order_density['medium'] = fuzz.trapmf(order_density.universe, [2, 4, 6, 8])
order_density['high'] = fuzz.trimf(order_density.universe, [6, 10, 10])

deliver_urgency['low'] = fuzz.trimf(deliver_urgency.universe, [0, 0, 4])
deliver_urgency['medium'] = fuzz.trapmf(deliver_urgency.universe, [2, 4, 6, 8])
deliver_urgency['high'] = fuzz.trimf(deliver_urgency.universe, [6, 10, 10])

driver_load['low'] = fuzz.trimf(driver_load.universe, [0, 0, 4])
driver_load['medium'] = fuzz.trapmf(driver_load.universe, [2, 4, 6, 8])
driver_load['high'] = fuzz.trimf(driver_load.universe, [6, 10, 10])

traffic_condition['low'] = fuzz.trimf(traffic_condition.universe, [0, 0, 4])
traffic_condition['medium'] = fuzz.trapmf(traffic_condition.universe, [2, 4, 6, 8])
traffic_condition['high'] = fuzz.trimf(traffic_condition.universe, [6, 10, 10])

profit_per_delivery['low'] = fuzz.trimf(profit_per_delivery.universe, [0, 0, 4])
profit_per_delivery['medium'] = fuzz.trapmf(profit_per_delivery.universe, [2, 4, 6, 8])
profit_per_delivery['high'] = fuzz.trimf(profit_per_delivery.universe, [6, 10, 10])

combine_orders['low'] = fuzz.trimf(combine_orders.universe, [0, 1, 4])
combine_orders['medium'] = fuzz.trimf(combine_orders.universe, [3, 5, 7])
combine_orders['high'] = fuzz.trimf(combine_orders.universe, [6, 9, 10])

delivery_priority['low'] = fuzz.trimf(delivery_priority.universe, [0, 1, 4])
delivery_priority['medium'] = fuzz.trimf(delivery_priority.universe, [3, 5, 7])
delivery_priority['high'] = fuzz.trimf(delivery_priority.universe, [6, 9, 10])

rule1 = ctrl.Rule(order_density['high'] & driver_load['low'] & traffic_condition['low'], combine_orders['high'])
rule2 = ctrl.Rule(order_density['medium'] & traffic_condition['high'] & deliver_urgency['medium'], combine_orders['medium'])
rule3 = ctrl.Rule(driver_load['high'] & order_density['high'] & profit_per_delivery['medium'], combine_orders['medium'])
rule4 = ctrl.Rule(order_density['low'] & deliver_urgency['high'] & traffic_condition['medium'], combine_orders['medium'])
rule5 = ctrl.Rule(profit_per_delivery['high'] & deliver_urgency['high'] & traffic_condition['high'], combine_orders['medium'])
rule6 = ctrl.Rule(deliver_urgency['high'] & profit_per_delivery['high'], delivery_priority['high'])
rule7 = ctrl.Rule(deliver_urgency['medium'] & traffic_condition['medium'], delivery_priority['medium'])
rule8 = ctrl.Rule(deliver_urgency['low'] & order_density['high'] & profit_per_delivery['low'], delivery_priority['low'])

logistics_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8])
logistics_sim = ctrl.ControlSystemSimulation(logistics_ctrl)

logistics_sim.input['order_density'] = 9.0
logistics_sim.input['driver_load'] = 2.0
logistics_sim.input['traffic_condition'] = 2.0
logistics_sim.input['deliver_urgency'] = 9.0
logistics_sim.input['profit_per_delivery'] = 9.0
logistics_sim.compute()

print(f"Chỉ số kết hợp đơn hàng: {logistics_sim.output['combine_orders']:.2f}")
print(f"Chỉ số ưu tiên giao hàng: {logistics_sim.output['delivery_priority']:.2f}")
```
