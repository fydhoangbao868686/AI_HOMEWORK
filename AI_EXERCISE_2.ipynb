Phiên bản này đã được chỉnh nhẹ một số khoảng giá trị mờ và dữ liệu đầu vào để số liệu khác hơn so với bản gốc, nhưng vẫn giữ nguyên ý tưởng xử lý của từng bài.

## Cài đặt và import thư viện

```python
%pip install -q scikit-fuzzy

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

## Bài 2.11

```python
# Bài 2.11 - Tính giá chuyến xe và điểm tài xế
Ride_distance = ctrl.Antecedent(np.arange(0, 51, 1), 'Ride_distance')
Traffic_condition = ctrl.Antecedent(np.arange(0, 101, 1), 'Traffic_condition')
Demand_level = ctrl.Antecedent(np.arange(0, 101, 1), 'Demand_level')
Weather_condition = ctrl.Antecedent(np.arange(0, 3.1, 0.1), 'Weather_condition')
Customer_rating = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'Customer_rating')
Ride_punctuality = ctrl.Antecedent(np.arange(0, 101, 1), 'Ride_punctuality')
Customer_fare = ctrl.Consequent(np.arange(0, 1010000, 1000), 'Customer_fare')
Driver_point = ctrl.Consequent(np.arange(0, 100, 1), 'Driver_point')

Ride_distance['S'] = fuzz.trimf(Ride_distance.universe, [0, 1.8, 3.4])
Ride_distance['M'] = fuzz.trimf(Ride_distance.universe, [2.4, 5.6, 8.8])
Ride_distance['L'] = fuzz.trimf(Ride_distance.universe, [6.8, 13.5, 21])
Ride_distance['VL'] = fuzz.trimf(Ride_distance.universe, [16, 33, 50])

Traffic_condition['L'] = fuzz.trimf(Traffic_condition.universe, [0, 18, 34])
Traffic_condition['M'] = fuzz.trimf(Traffic_condition.universe, [24, 52, 74])
Traffic_condition['H'] = fuzz.trimf(Traffic_condition.universe, [64, 88, 100])

Demand_level['L'] = fuzz.trimf(Demand_level.universe, [0, 18, 34])
Demand_level['M'] = fuzz.trimf(Demand_level.universe, [24, 53, 76])
Demand_level['H'] = fuzz.trimf(Demand_level.universe, [64, 92, 100])

Weather_condition['Good'] = fuzz.trimf(Weather_condition.universe, [0, 0.9, 1.4])
Weather_condition['Moderate'] = fuzz.trimf(Weather_condition.universe, [1.1, 2.0, 2.6])
Weather_condition['Bad'] = fuzz.trimf(Weather_condition.universe, [2.2, 2.8, 3])

Customer_rating['L'] = fuzz.trimf(Customer_rating.universe, [0, 1.2, 2.2])
Customer_rating['M'] = fuzz.trimf(Customer_rating.universe, [1.8, 3.0, 4.1])
Customer_rating['H'] = fuzz.trimf(Customer_rating.universe, [3.9, 4.8, 5])

Ride_punctuality['Late'] = fuzz.trimf(Ride_punctuality.universe, [0, 28, 52])
Ride_punctuality['On_time'] = fuzz.trimf(Ride_punctuality.universe, [42, 62, 82])
Ride_punctuality['Early'] = fuzz.trimf(Ride_punctuality.universe, [72, 96, 100])

Customer_fare['Low'] = fuzz.trimf(Customer_fare.universe, [0, 18000, 30000])
Customer_fare['Medium'] = fuzz.trimf(Customer_fare.universe, [28000, 42000, 56000])
Customer_fare['High'] = fuzz.trimf(Customer_fare.universe, [54000, 70000, 86000])
Customer_fare['Very High'] = fuzz.trimf(Customer_fare.universe, [82000, 105000, 105000])

Driver_point['None'] = fuzz.trimf(Driver_point.universe, [0, 8, 16])
Driver_point['Few'] = fuzz.trimf(Driver_point.universe, [14, 34, 52])
Driver_point['Moderate'] = fuzz.trimf(Driver_point.universe, [48, 66, 82])
Driver_point['High'] = fuzz.trimf(Driver_point.universe, [78, 94, 100])

rule1 = ctrl.Rule(Customer_rating['L'] & Ride_punctuality['Late'], Driver_point['None'])
rule2 = ctrl.Rule(Customer_rating['L'] & Ride_punctuality['On_time'], Driver_point['Few'])
rule3 = ctrl.Rule(Customer_rating['L'] & Ride_punctuality['Early'], Driver_point['Few'])
rule4 = ctrl.Rule(Customer_rating['M'] & Ride_punctuality['Late'], Driver_point['Few'])
rule5 = ctrl.Rule(Customer_rating['M'] & Ride_punctuality['On_time'], Driver_point['Moderate'])
rule6 = ctrl.Rule(Customer_rating['M'] & Ride_punctuality['Early'], Driver_point['Moderate'])
rule7 = ctrl.Rule(Customer_rating['H'] & Ride_punctuality['Late'], Driver_point['Moderate'])
rule8 = ctrl.Rule(Customer_rating['H'] & Ride_punctuality['On_time'], Driver_point['High'])
rule9 = ctrl.Rule(Customer_rating['H'] & Ride_punctuality['Early'], Driver_point['High'])

rule10 = ctrl.Rule(Ride_distance['S'] & Traffic_condition['L'] & Demand_level['L'], Customer_fare['Low'])
rule11 = ctrl.Rule(Ride_distance['S'] & Traffic_condition['M'] & Demand_level['H'], Customer_fare['Medium'])
rule12 = ctrl.Rule(Ride_distance['M'] & Traffic_condition['H'] & Demand_level['H'], Customer_fare['Very High'])
rule13 = ctrl.Rule(Ride_distance['L'] & Traffic_condition['M'] & Weather_condition['Good'], Customer_fare['High'])
rule14 = ctrl.Rule(Ride_distance['L'] & Traffic_condition['H'] & Weather_condition['Moderate'], Customer_fare['High'])
rule15 = ctrl.Rule(Ride_distance['VL'] & Traffic_condition['H'] & Demand_level['H'], Customer_fare['Very High'])
rule16 = ctrl.Rule(Ride_distance['M'] & Traffic_condition['L'] & Demand_level['L'], Customer_fare['Medium'])
rule17 = ctrl.Rule(Ride_distance['S'] & Traffic_condition['H'] & Weather_condition['Bad'], Customer_fare['High'])
rule18 = ctrl.Rule(Ride_distance['VL'] & Weather_condition['Bad'], Customer_fare['Very High'])
rule19 = ctrl.Rule(Ride_distance['M'] & Traffic_condition['M'] & Weather_condition['Moderate'], Customer_fare['Medium'])
rule20 = ctrl.Rule(Ride_distance['M'] & Traffic_condition['H'] & Demand_level['M'], Customer_fare['High'])

control_system = ctrl.ControlSystem([
    rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10,
    rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20
])

ride_system = ctrl.ControlSystemSimulation(control_system)

ride_system.input['Ride_distance'] = 31.8
ride_system.input['Traffic_condition'] = 87
ride_system.input['Demand_level'] = 92
ride_system.input['Weather_condition'] = 2.7
ride_system.input['Customer_rating'] = 4.8
ride_system.input['Ride_punctuality'] = 88

ride_system.compute()

print('Giá đi xe:', f"{ride_system.output['Customer_fare']:.2f}")
print('Điểm tài xế:', f"{ride_system.output['Driver_point']:.2f}")
```

## Bài 2.12

```python
# Bài 2.12 - Xác định phần trăm giảm giá theo thông tin cửa hàng
Store_rating = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'Store_rating')
Sales_volume = ctrl.Antecedent(np.arange(0, 10001, 1), 'Sales_volume')
Profit_margin = ctrl.Antecedent(np.arange(0, 101, 1), 'Profit_margin')
Seasonal_event = ctrl.Antecedent(np.arange(0, 11, 1), 'Seasonal_event')
Competitor_discounts = ctrl.Antecedent(np.arange(0, 71, 1), 'Competitor_discounts')
Discount_pct = ctrl.Consequent(np.arange(0, 71, 1), 'Discount_pct')

Store_rating['Low'] = fuzz.trimf(Store_rating.universe, [0, 2.1, 3.9])
Store_rating['Medium'] = fuzz.trimf(Store_rating.universe, [3.9, 4.3, 4.6])
Store_rating['High'] = fuzz.trimf(Store_rating.universe, [4.5, 4.9, 5])

Sales_volume['Low'] = fuzz.trimf(Sales_volume.universe, [0, 600, 1600])
Sales_volume['Medium'] = fuzz.trimf(Sales_volume.universe, [1200, 4200, 7200])
Sales_volume['High'] = fuzz.trimf(Sales_volume.universe, [6800, 9600, 10000])

Profit_margin['Low'] = fuzz.trimf(Profit_margin.universe, [0, 18, 38])
Profit_margin['Medium'] = fuzz.trimf(Profit_margin.universe, [28, 52, 76])
Profit_margin['High'] = fuzz.trimf(Profit_margin.universe, [66, 95, 100])

Seasonal_event['None'] = fuzz.trimf(Seasonal_event.universe, [0, 1.5, 2.5])
Seasonal_event['Moderate'] = fuzz.trimf(Seasonal_event.universe, [1.5, 5.5, 7])
Seasonal_event['High'] = fuzz.trimf(Seasonal_event.universe, [6, 9.5, 10])

Competitor_discounts['Low'] = fuzz.trimf(Competitor_discounts.universe, [0, 14, 32])
Competitor_discounts['Medium'] = fuzz.trimf(Competitor_discounts.universe, [26, 38, 52])
Competitor_discounts['High'] = fuzz.trimf(Competitor_discounts.universe, [43, 58, 70])

Discount_pct['Very Low'] = fuzz.trimf(Discount_pct.universe, [0, 4, 6])
Discount_pct['Low'] = fuzz.trimf(Discount_pct.universe, [5, 11, 14])
Discount_pct['Medium'] = fuzz.trimf(Discount_pct.universe, [12, 22, 28])
Discount_pct['High'] = fuzz.trimf(Discount_pct.universe, [24, 42, 48])
Discount_pct['Very High'] = fuzz.trimf(Discount_pct.universe, [42, 66, 70])

rule1 = ctrl.Rule(Store_rating['High'] & Sales_volume['High'] & Profit_margin['High'], Discount_pct['Very Low'])
rule2 = ctrl.Rule(Store_rating['Low'] & Sales_volume['Low'] & Profit_margin['High'], Discount_pct['High'])
rule3 = ctrl.Rule(Seasonal_event['High'] & Competitor_discounts['High'], Discount_pct['Very High'])
rule4 = ctrl.Rule(Store_rating['Medium'] & Sales_volume['Medium'] & Profit_margin['Medium'], Discount_pct['Medium'])
rule5 = ctrl.Rule(Competitor_discounts['Low'] & Profit_margin['Low'] & Sales_volume['High'], Discount_pct['Very Low'])
rule6 = ctrl.Rule(Store_rating['Low'] & Seasonal_event['None'], Discount_pct['Medium'])
rule7 = ctrl.Rule(Sales_volume['Low'] & Profit_margin['Low'], Discount_pct['Very High'])

rule_control = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
discount_system = ctrl.ControlSystemSimulation(rule_control)

discount_system.input['Store_rating'] = 4.35
discount_system.input['Sales_volume'] = 5200
discount_system.input['Profit_margin'] = 22
discount_system.input['Seasonal_event'] = 8
discount_system.input['Competitor_discounts'] = 64

discount_system.compute()

print("Phần trăm giảm giá là:", f"{discount_system.output['Discount_pct']:.2f}")
```

## Bài 2.13

```python
# Bài 2.13 - Đề xuất mức giảm giá sản phẩm
PD = ctrl.Antecedent(np.arange(0, 101, 1), 'Product_Demand')
CPP = ctrl.Antecedent(np.arange(0, 11, 1), 'Competitor_pricing_pressure')
SR = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'Store_Reputation')
PM = ctrl.Antecedent(np.arange(0, 101, 1), 'Profit_margin')
SD = ctrl.Antecedent(np.arange(0, 11, 1), 'Seasonal_demand')
Dis_pct = ctrl.Consequent(np.arange(0, 71, 1), 'Discount_pct')

PD['Low'] = fuzz.trimf(PD.universe, [0, 22, 32])
PD['Medium'] = fuzz.trimf(PD.universe, [22, 52, 72])
PD['High'] = fuzz.trapmf(PD.universe, [62, 82, 98, 100])

CPP['Low'] = fuzz.trimf(CPP.universe, [0, 1.2, 2])
CPP['Medium'] = fuzz.trapmf(CPP.universe, [1.5, 2.5, 5, 7.5])
CPP['High'] = fuzz.trimf(CPP.universe, [6.5, 9, 10])

SR['Low'] = fuzz.trimf(SR.universe, [0, 1.4, 3.8])
SR['Medium'] = fuzz.trapmf(SR.universe, [3.7, 4.0, 4.5, 4.7])
SR['High'] = fuzz.trimf(SR.universe, [4.5, 4.9, 5])

PM['Low'] = fuzz.trimf(PM.universe, [0, 12, 28])
PM['Medium'] = fuzz.trimf(PM.universe, [22, 34, 46])
PM['High'] = fuzz.trimf(PM.universe, [34, 62, 72])

SD['Low'] = fuzz.trimf(SD.universe, [0, 1.2, 2.4])
SD['Medium'] = fuzz.trapmf(SD.universe, [1.8, 4.2, 5.6, 7.2])
SD['High'] = fuzz.trimf(SD.universe, [6.2, 8.8, 10])

Dis_pct['Very Low'] = fuzz.trimf(Dis_pct.universe, [0, 4, 6])
Dis_pct['Low'] = fuzz.trapmf(Dis_pct.universe, [4, 6, 10, 12])
Dis_pct['Medium'] = fuzz.trapmf(Dis_pct.universe, [10, 13, 21, 24])
Dis_pct['High'] = fuzz.trapmf(Dis_pct.universe, [22, 28, 43, 58])
Dis_pct['Very High'] = fuzz.trimf(Dis_pct.universe, [52, 66, 70])

rule1 = ctrl.Rule(PD['High'] & CPP['Low'] & PM['Low'], Dis_pct['Very Low'])
rule2 = ctrl.Rule(PD['Low'] & CPP['High'] & PM['High'], Dis_pct['High'])
rule3 = ctrl.Rule(SR['High'] & PM['Medium'] & SD['High'], Dis_pct['Medium'])
rule4 = ctrl.Rule(CPP['High'] & SD['High'] & PM['High'], Dis_pct['Very High'])
rule5 = ctrl.Rule(SR['Low'] & PD['Medium'] & PM['Low'], Dis_pct['Medium'])
rule6 = ctrl.Rule(PD['High'] & SD['Low'] & CPP['Low'], Dis_pct['Very Low'])
rule7 = ctrl.Rule(PM['High'] & CPP['Medium'], Dis_pct['Medium'])

rule_control = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
discount_percentage = ctrl.ControlSystemSimulation(rule_control)

discount_percentage.input['Product_Demand'] = 78
discount_percentage.input['Competitor_pricing_pressure'] = 4.5
discount_percentage.input['Store_Reputation'] = 4.3
discount_percentage.input['Profit_margin'] = 58
discount_percentage.input['Seasonal_demand'] = 8.5

discount_percentage.compute()

print("Phần trăm giảm giá là:", f"{discount_percentage.output['Discount_pct']:.2f}")
```

## Bài 2.14

```python
# Bài 2.14 - Đánh giá kết hợp đơn hàng và độ ưu tiên giao hàng
order_density = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'order_density')
deliver_urgency = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'deliver_urgency')
driver_load = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'driver_load')
traffic_condition = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'traffic_condition')
profit_per_delivery = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'profit_per_delivery')

combine_orders = ctrl.Consequent(np.arange(0, 11, 1), 'combine_orders')
delivery_priority = ctrl.Consequent(np.arange(0, 11, 1), 'delivery_priority')

order_density['low'] = fuzz.trimf(order_density.universe, [0, 0.8, 4.2])
order_density['medium'] = fuzz.trapmf(order_density.universe, [2.2, 4.2, 6.2, 8.2])
order_density['high'] = fuzz.trimf(order_density.universe, [6.2, 9.4, 10])

deliver_urgency['low'] = fuzz.trimf(deliver_urgency.universe, [0, 0.7, 3.8])
deliver_urgency['medium'] = fuzz.trapmf(deliver_urgency.universe, [2.0, 4.0, 6.3, 8.0])
deliver_urgency['high'] = fuzz.trimf(deliver_urgency.universe, [6.0, 9.3, 10])

driver_load['low'] = fuzz.trimf(driver_load.universe, [0, 1.0, 4.4])
driver_load['medium'] = fuzz.trapmf(driver_load.universe, [2.3, 4.3, 6.1, 8.1])
driver_load['high'] = fuzz.trimf(driver_load.universe, [6.3, 9.5, 10])

traffic_condition['low'] = fuzz.trimf(traffic_condition.universe, [0, 0.9, 4.0])
traffic_condition['medium'] = fuzz.trapmf(traffic_condition.universe, [2.1, 4.1, 6.1, 8.1])
traffic_condition['high'] = fuzz.trimf(traffic_condition.universe, [6.1, 9.4, 10])

profit_per_delivery['low'] = fuzz.trimf(profit_per_delivery.universe, [0, 0.8, 3.9])
profit_per_delivery['medium'] = fuzz.trapmf(profit_per_delivery.universe, [2.1, 4.2, 6.2, 8.2])
profit_per_delivery['high'] = fuzz.trimf(profit_per_delivery.universe, [6.2, 9.2, 10])

combine_orders['low'] = fuzz.trimf(combine_orders.universe, [0, 1.2, 4.2])
combine_orders['medium'] = fuzz.trimf(combine_orders.universe, [3.2, 5.2, 7.2])
combine_orders['high'] = fuzz.trimf(combine_orders.universe, [6.2, 9.0, 10])

delivery_priority['low'] = fuzz.trimf(delivery_priority.universe, [0, 1.2, 4.2])
delivery_priority['medium'] = fuzz.trimf(delivery_priority.universe, [3.2, 5.1, 7.1])
delivery_priority['high'] = fuzz.trimf(delivery_priority.universe, [6.1, 9.1, 10])

rule1 = ctrl.Rule(order_density['high'] & driver_load['low'] & traffic_condition['low'], combine_orders['high'])
rule2 = ctrl.Rule(order_density['medium'] & traffic_condition['high'] & deliver_urgency['medium'], combine_orders['medium'])
rule3 = ctrl.Rule(driver_load['high'] & order_density['high'] & profit_per_delivery['medium'], combine_orders['medium'])
rule4 = ctrl.Rule(order_density['low'] & deliver_urgency['high'] & traffic_condition['medium'], combine_orders['medium'])
rule5 = ctrl.Rule(profit_per_delivery['high'] & deliver_urgency['high'] & traffic_condition['high'], combine_orders['medium'])
rule6 = ctrl.Rule(deliver_urgency['high'] & profit_per_delivery['high'], delivery_priority['high'])
rule7 = ctrl.Rule(deliver_urgency['medium'] & traffic_condition['medium'], delivery_priority['medium'])
rule8 = ctrl.Rule(deliver_urgency['low'] & order_density['high'] & profit_per_delivery['low'], delivery_priority['low'])

logistics_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8])
logistics_sim = ctrl.ControlSystemSimulation(logistics_ctrl)

logistics_sim.input['order_density'] = 8.8
logistics_sim.input['driver_load'] = 2.2
logistics_sim.input['traffic_condition'] = 2.3
logistics_sim.input['deliver_urgency'] = 8.7
logistics_sim.input['profit_per_delivery'] = 8.5

logistics_sim.compute()

print(f"Chỉ số kết hợp đơn hàng: {logistics_sim.output['combine_orders']:.2f}")
print(f"Chỉ số ưu tiên giao hàng: {logistics_sim.output['delivery_priority']:.2f}")
```
